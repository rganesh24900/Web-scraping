```python
import requests
import re
```

```python
header={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
a=requests.get('https://www.ndtv.com/india-news/40-arrested-after-protests-inside-srinagars-jamia-masjid-2387674',headers=header,timeout=60)
```

```python
b=a.text
```

```python
b
```

```python
headline=re.search('property="taboola:title"\s*content="(.*?)"',b,re.S).group(1)                  
```

```python
headline
```

```python
date=re.search('"publicationdate":"(.*?)"',b,flags=re.S).group(1)
```

```python
date
```

```python
content=re.search('<p class="ins_instory_dv_caption\s*sp_b">(.*?)</p>',b,flags=re.S).group(1)
content=re.sub('<.*?>','',b,flags=re.S)
```

```python
author=re.search('"author":"(.*?)"',b,flags=re.S).group(1)
```
